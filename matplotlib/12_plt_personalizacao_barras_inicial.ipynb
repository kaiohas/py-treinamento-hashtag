## Gráfico de barras para apresentação

Considere que você tem as quantidades de produtos vendidos em uma loja. Faça um gráfico de barras horizontais.

```python
import matplotlib.pyplot as plt

produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]
```

```python
plt.figure(figsize=(15,5))

plt.bar(produtos,quantidades)

plt.show()
```

```python
plt.figure(figsize=(15,5))

barras = plt.barh(produtos,quantidades)
plt.bar_label(barras,fontsize=10,weight='bold',padding=-55,color='white')
plt.xticks([])
plt.tick_params(axis='y',length=0)
plt.title('Quantidade vendida por produto')
plt.box()
plt.show()
```

```python
from cycler import cycler
plt.figure(figsize=(15,5))
cores = plt.get_cmap('Dark2').colors
ciclo_cores = cycler('color',cores)
plt.rc('axes',prop_cycle=ciclo_cores)

rotulos = [valor if valor == max(quantidades) else '' for valor in quantidades]
barras = plt.barh(produtos,quantidades)
plt.bar_label(barras,labels=rotulos,fontsize=10,weight='bold',padding=-55,color='white')
plt.xticks([])
plt.tick_params(axis='y',length=0)



for i,barra in enumerate(barras):
    if i != quantidades.index(max(quantidades)):
        barras[i].set_color(cores[1])
        barras[i].set_alpha(0.5)




plt.title('Quantidade vendida por produto')
plt.box()
plt.show()
```

```python
from cycler import cycler
plt.figure(figsize=(15,5))
cores = plt.get_cmap('Dark2').colors
ciclo_cores = cycler('color',cores)
plt.rc('axes',prop_cycle=ciclo_cores)

rotulos = [valor if valor == max(quantidades_ordenadas) else '' for valor in quantidades_ordenadas]
barras = plt.bar(produtos_ordenados,quantidades_ordenadas)
plt.bar_label(barras,labels=rotulos,fontsize=10,weight='bold',padding=-55,color='white')
plt.yticks([])
plt.tick_params(axis='x',length=0)



for i,barra in enumerate(barras):
    if i != quantidades_ordenadas.index(max(quantidades_ordenadas)):
        barras[i].set_color(cores[1])
        barras[i].set_alpha(0.5)


plt.gca().invert_xaxis()

plt.title('Quantidade vendida por produto')
plt.box()
plt.show()
```
