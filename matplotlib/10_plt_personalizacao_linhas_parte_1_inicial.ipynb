## Gráfico de linha para apresentação


```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
#plot(x,y)
plt.plot(meses, vendas,marker='o',markersize=5,linestyle='--',linewidth=2)
plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
plt.plot(meses, vendas,marker='o')
plt.annotate('Menor Valor',color='red',xy=(2.1,78),xytext=(3,80),arrowprops=dict(facecolor='red'))
plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
import numpy as np
from cycler import cycler



#Falando de cores
cores = plt.get_cmap('Dark2').colors
ciclo_cores = cycler('color',cores)
plt.rc('axes',prop_cycle=ciclo_cores)

#Criando marcadores
marcadores = np.where(vendas == np.min(vendas), 'x','')
for i,valor in enumerate(marcadores):
    if valor == 'x':
        cor=cores[1]
        tamanho = 80
    else:
        cor=cores[2]
        tamanho = 0
    plt.annotate(vendas[i],color=cor,xy=(meses[i],vendas[i]),xytext=(-10,10),textcoords='offset points',weight='bold')
    plt.scatter(meses[i],vendas[i],marker=valor,color=cor,s=tamanho)

#Plotando e anotações
plt.plot(meses, vendas)
#plt.annotate('Menor Valor',color='red',xy=(2.2,78),xytext=(3,78),arrowprops=dict(facecolor='red'))

#Destaque Regiões
plt.axvspan(6,11,alpha=0.2,color='red')
plt.text(6,85,'Recuperação de crescimento',color='green',fontsize=9,weight='bold')




plt.box(False)
plt.tick_params(axis='x',length=0)
plt.yticks([])

plt.grid(axis='x',linestyle='--',alpha=0.5)
#plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()

```
