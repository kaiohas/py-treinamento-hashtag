# Trabalhando com eixos



## Mais de um eixo y em uma figura

Crie uma figura de um gráfico de precipitação com um eixo secundário para a temperatura, para mostrar a relação entre a temperatura e a precipitação em uma cidade.



```python
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
precipitacao_media = [200, 180, 160, 100, 60, 30, 30, 35, 60, 100, 160, 200]
temperatura_media = [30, 29, 27, 25, 23, 22, 22, 23, 24, 25, 26, 28]
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()
cor_temperatura  = 'tab:red'
cor_precipitacao  = 'tab:blue'

ax1.bar(meses,precipitacao_media,color=cor_precipitacao,label='Precipitacao media')
ax1.set_ylabel('Precipitacao media',color=cor_precipitacao)
ax1.tick_params(axis='y',labelcolor=cor_precipitacao)
ax2 = ax1.twinx()
ax1.set_ylim(0,220)
ax1.tick_params(axis='x',length=0)

ax2.plot(meses,temperatura_media,color=cor_temperatura,linewidth=2,label='Temperatura media')
ax2.set_ylabel('Temperatura media',color=cor_temperatura)
ax2.tick_params(axis='y',labelcolor=cor_temperatura)
ax2.set_ylim(20,31)

ax1.set_title('Temperatura e precipitação ao longo do ano')

fig.legend(bbox_to_anchor=(0.75,0.95),bbox_transform=ax1.transAxes)

plt.show()
```
