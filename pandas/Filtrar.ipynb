```python
import pandas as pd

tabela = pd.read_csv(r'exportacoes_franca.csv')
tabela.head()
```

```python
tabela_anos = tabela[['Year','US$ FOB']].groupby('Year').sum()

def formatar(valor):
    valor_formatado = f"US${valor:,}"
    return valor_formatado

tabela_anos['US$ FOB'] = tabela_anos['US$ FOB'].apply(formatar)

tabela_anos
```

```python
tabela.groupby('Year').sum(numeric_only=True)
```

```python
ranking_produtos = tabela[['SH4 Description','US$ FOB']].groupby('SH4 Description').sum(numeric_only=True)

ranking_produtos = ranking_produtos.sort_values(by='US$ FOB',ascending=False)
ranking_produtos["US$ FOB"] = ranking_produtos['US$ FOB'].apply(formatar)
ranking_produtos[:11]
```

```python
tabela_2020 = tabela.loc[tabela['Year']==2020,:]
tabela_2020.head()
```

```python
tabela_2020_cidade = tabela_2020[['City','US$ FOB']].groupby('City').sum()
tabela_2020_cidade = tabela_2020_cidade.sort_values(by='US$ FOB',ascending=False)
tabela_2020_cidade["US$ FOB"] = tabela_2020_cidade["US$ FOB"].apply(formatar)

tabela_2020_cidade[:4].index[1]
```

```python
cidade = tabela_2020_cidade[:4].index[2]

tabela_cidade = tabela_2020.loc[tabela_2020['City']==cidade,:]

tabela_produtdo_cidade = tabela_2020[['SH2 Description','US$ FOB']].groupby('SH2 Description').sum()
tabela_produtdo_cidade = tabela_produtdo_cidade.sort_values(by='US$ FOB',ascending=False)
tabela_produtdo_cidade['US$ FOB'] = tabela_produtdo_cidade['US$ FOB'].apply(formatar)
display(cidade)
tabela_produtdo_cidade
```
