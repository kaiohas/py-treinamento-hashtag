```python
tabela_final = pd.DataFrame()
pular_indice = 0
while True:
    
    link = fr"https://olinda.bcb.gov.br/olinda/servico/mecir_dinheiro_em_circulacao/versao/v1/odata/informacoes_diarias?$top=10000&$skip={pular_indice}&$orderby=Data%20desc&$format=json"
    requisicao = requests.get(link)
    info = requisicao.json()
    tabela = pd.DataFrame(info['value'])
    if len(info['value'])<1:
        break
    tabela_final = pd.concat([tabela_final,tabela])
    pular_indice+=10000

tabela_final
```
