#### Pegando cotações semanais - JSON

```python
from chave_alpha import chave_api
import pprint
import requests
link = f'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY_ADJUSTED&symbol=CA.PA&apikey={chave_api}'
site = requests.get(link)
data = site.json()

pprint.pprint(data)
```

#### Pegando cotações semanais - CSV

```python
import pandas as pd
from io import StringIO
link = f'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY_ADJUSTED&symbol=CA.PA&apikey={chave_api}&datatype=csv'

site = requests.get(link)
data = pd.read_csv(StringIO(site.text))

data
```

### Cotações

```python
acoes = ['ITUB4', 'ABEV3', 'BBAS3']
codigo = acoes[1]
tabela_fim = pd.DataFrame()
for codigo in acoes:
    link = f'https://www.alphavantage.co/query?function=GLOBAL_QUOTE&symbol={codigo}.SAO&apikey={chave_api}&datatype=csv'
    site = requests.get(link)
    data = pd.read_csv(StringIO(site.text))
    tabelas = [tabela_fim,data]
    tabela_fim = pd.concat(tabelas)


tabela_fim
```
